<h4> 2. Pour obtenir la table T des villes du monde, on fusionne les fichiers ville_i.csv en exécutant le code suivant </h4>

```python
M = []
for i in range(1,13):
    M.append(pa.read_csv('villes_{}.csv'.format(i)))
T = pa.concat(M,ignore_index=True,axis=0)
T.head() 
```
